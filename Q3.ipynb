```python
import pandas as pd
import numpy as np
import datetime

```

```python
df = pd.read_csv("/home/spark/Downloads/archive/rejected_2007_to_2018Q4.csv")
```

```python
# remove spaces in columns name
df.columns = df.columns.str.replace(' ', '')
df.columns
```

```python
 # Filter data onwards 2009
#df['ApplicationDate'] = pd.to_datetime(df['ApplicationDate'], errors='coerce')
#df[df['ApplicationDate'].dt.strftime('%Y-%m-%d') > '2009-12-31']
mask = (df['ApplicationDate'] > '2009-12-31')
df = df.loc[mask]
```

```python
def MA50():
    # Filter data onwards 2009

    df['MA50'] = df['Risk_Score'].rolling(50).mean()
 
    # removing all the NULL values using
    df.dropna(inplace=True)
    
    # printing Dataframe
    return df
    df.head(100)
MA50()
```

```python
def MA100():
   
    df['MA100'] = df['Risk_Score'].rolling(100).mean()
 
    # removing all the NULL values using
    df.dropna(inplace=True)

    # printing Dataframe
    return df
    df.head(100)
MA100()
```
