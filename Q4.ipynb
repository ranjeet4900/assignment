```python
import pandas as pd
import numpy as np
import datetime

```

```python
df = pd.read_csv("/home/spark/Downloads/archive/rejected_2007_to_2018Q4.csv")
```

```python
# remove spaces in columns name
df.columns = df.columns.str.replace(' ', '')
df.columns
```

```python
def MA50():
    # Filter data onwards 2009

    df['MA50'] = df['Risk_Score'].rolling(50).mean()
 
    # removing all the NULL values using
    df.dropna(inplace=True)
    
    # printing Dataframe
    return df
    df.head(100)
MA50()
```

```python
## By using MA50 column calculating EMA50
def EMA50():
    # Filter data onwards 2009

    df['MA50'] = df['Risk_Score'].ewm(span=50).mean()
    
    # removing all the NULL values using
    df.dropna(inplace=True)
    
    # printing Dataframe
    return df
    df.head(100)
EMA50()
```
